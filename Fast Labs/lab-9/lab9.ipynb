```python
import numpy as np

# Load CSV data into a NumPy array
data = np.genfromtxt('stock_prices.csv', delimiter=',', skip_header=1)

# Set print options to display numbers with 2 decimal places
np.set_printoptions(precision=2, suppress=True)

# Now when you print the array, it will display numbers with 2 decimal places and no scientific notation
print(data)
```

```python
# Get the "Close" column (assumed to be the 5th column)
close_prices = data[:, 4]

# Calculate the daily returns
daily_returns = (close_prices[1:] - close_prices[:-1]) / close_prices[:-1]

print(daily_returns)
```

```python
# Get the "Volume" column (assumed to be the 6th column)
volume = data[:, 5]

# Find the index of the maximum volume
max_volume_index = np.argmax(volume)

# Get the corresponding date (assumed to be the 1st column)
date_of_max_volume = data[max_volume_index, 0]

print(f"Date with the highest trading volume: {date_of_max_volume}")
```

```python
import datetime

# Get the "Date" column (assumed to be the 1st column)
dates = data[:, 0]
closing_prices = data[:, 4]

# Convert the dates to actual datetime objects
dates = [datetime.datetime.strptime(str(int(date)), '%Y%m%d').date() for date in dates]

# Identify the days of the week for Mondays and Fridays
mondays = []
fridays = []
closing_prices_monday = []
closing_prices_friday = []

for i, date in enumerate(dates):
    day_of_week = date.weekday()  # Monday = 0, Sunday = 6
    if day_of_week == 0:  # Monday
        mondays.append(i)
        closing_prices_monday.append(closing_prices[i])
    elif day_of_week == 4:  # Friday
        fridays.append(i)
        closing_prices_friday.append(closing_prices[i])

# Calculate the average closing prices for Mondays and Fridays
avg_closing_price_monday = np.mean(closing_prices_monday)
avg_closing_price_friday = np.mean(closing_prices_friday)

print(f"Average closing price on Mondays: {avg_closing_price_monday}")
print(f"Average closing price on Fridays: {avg_closing_price_friday}")
```

```python
import numpy as np
import pandas as pd

# Generate random data for temperature measurements
years = np.array([2020, 2021, 2022, 2023, 2024] * 2)  # 5 years, repeated for 10 locations
locations = np.arange(1, 11)  # 10 locations
temperature_data = np.random.uniform(-30, 50, (10, 5))  # Random temperatures between -30°C to 50°C for each year

# Combine the years, locations, and temperature data into a DataFrame
data = pd.DataFrame(temperature_data, columns=[f"Year_{year}" for year in years[:5]], index=locations)

# Display the first few rows of the generated data
print(data)
```

```python
# Calculate mean, standard deviation, and variance for the temperature data
mean_temp = np.mean(data.values)  # Mean of all temperature values in the dataset
std_temp = np.std(data.values)    # Standard deviation of all temperature values
var_temp = np.var(data.values)    # Variance of all temperature values

# Print the results
print(f"Mean Temperature: {mean_temp}")
print(f"Standard Deviation: {std_temp}")
print(f"Variance: {var_temp}")
```

```python
# Calculate the average temperature for each location (across all years)
location_avg_temp = np.mean(data.values, axis=1)  # Average across each row (location)

# Identify the location with the highest and lowest average temperature
max_temp_index = np.argmax(location_avg_temp)  # Location with highest avg temp
min_temp_index = np.argmin(location_avg_temp)  # Location with lowest avg temp

print(f"Location with the highest average temperature: {max_temp_index}")
print(f"Location with the lowest average temperature: {min_temp_index}")
```

```python
# Calculate variance of temperatures for each year
yearly_variances = np.var(data.values, axis=0)  # Variance along columns (years)

# Find the year with the highest variance
max_variance_year = np.argmax(yearly_variances)  # Year with highest temperature variance

# Print the result
print(f"The year with the highest temperature variance: {years[max_variance_year]}")
```

```python
import numpy as np
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt

# Creating sample data for demonstration
# First, let's create some sample stock data
dates = pd.date_range(start='2023-01-01', end='2023-01-31')
np.random.seed(42)  # For reproducible results
open_prices = np.random.uniform(100, 110, len(dates))
high_prices = open_prices + np.random.uniform(0, 5, len(dates))
low_prices = open_prices - np.random.uniform(0, 5, len(dates))
close_prices = open_prices + np.random.uniform(-3, 3, len(dates))
volumes = np.random.randint(1000, 10000, len(dates))

# Creating a DataFrame and saving it to CSV for demonstration
stock_data = pd.DataFrame({
    'Date': dates,
    'Open': open_prices,
    'High': high_prices,
    'Low': low_prices,
    'Close': close_prices,
    'Volume': volumes
})
stock_data.to_csv('stock_data.csv', index=False)
print("Sample stock data created and saved to 'stock_data.csv'")

# Now let's go through each task

print("\n--- STOCK DATA ANALYSIS ---\n")

# 1. Load the CSV data into a NumPy array
print("1. Loading CSV data into NumPy array:")
# Using pandas to load and then convert to numpy
df = pd.read_csv('stock_data.csv')
# Convert to structured numpy array to keep column names
stock_array = df.to_numpy()
date_strings = df['Date'].values
# For numerical calculations, we'll use just the numerical columns
numeric_data = df[['Open', 'High', 'Low', 'Close', 'Volume']].to_numpy()

print(f"Shape of numeric data: {numeric_data.shape}")
print(f"First few rows:\n{numeric_data[:3]}")

# 2. Calculate the daily returns for the "Close" prices
print("\n2. Calculating daily returns for 'Close' prices:")
close_prices = numeric_data[:, 3]  # 4th column (0-indexed)
daily_returns = np.diff(close_prices) / close_prices[:-1] * 100  # Percentage returns

print(f"Close prices: {close_prices[:5]}")
print(f"Daily returns (%): {daily_returns[:5]}")

# 3. Find the date on which the stock had the highest trading volume
print("\n3. Finding date with highest trading volume:")
volumes = numeric_data[:, 4]  # 5th column (0-indexed)
max_volume_idx = np.argmax(volumes)
max_volume_date = date_strings[max_volume_idx]
max_volume = volumes[max_volume_idx]

print(f"Highest volume: {max_volume} on {max_volume_date}")

# 4. Determine the average closing price for Mondays and Fridays
print("\n4. Average closing price for Mondays and Fridays:")
# Convert string dates to datetime objects
dates = pd.to_datetime(date_strings)
# Get weekday (0=Monday, 4=Friday)
weekdays = np.array([d.weekday() for d in dates])

monday_mask = (weekdays == 0)
friday_mask = (weekdays == 4)

monday_closes = close_prices[monday_mask]
friday_closes = close_prices[friday_mask]

monday_avg = np.mean(monday_closes) if len(monday_closes) > 0 else np.nan
friday_avg = np.mean(friday_closes) if len(friday_closes) > 0 else np.nan

print(f"Average Monday closing price: {monday_avg:.2f}")
print(f"Average Friday closing price: {friday_avg:.2f}")

# 5. Identify the top three days with the largest intraday price fluctuations
print("\n5. Top three days with largest intraday fluctuations:")
high_prices = numeric_data[:, 1]  # 2nd column (0-indexed)
low_prices = numeric_data[:, 2]   # 3rd column (0-indexed)

# Calculate the price range (High - Low)
price_range = high_prices - low_prices
# Get indices of top three ranges
top_three_idx = np.argsort(price_range)[-3:][::-1]  # Sort and get last 3, then reverse

print("Date\t\tHigh\tLow\tRange")
for idx in top_three_idx:
    print(f"{date_strings[idx]}\t{high_prices[idx]:.2f}\t{low_prices[idx]:.2f}\t{price_range[idx]:.2f}")

print("\n--- TEMPERATURE DATA ANALYSIS ---\n")

# Creating sample temperature data
# We'll create a 3D array: [year, location, temperature]
years = 3
locations = 4
measurements_per_year = 12  # monthly measurements

np.random.seed(42)
# Shape: (years, locations, measurements)
temp_data = np.random.normal(15, 5, (years, locations, measurements_per_year))

# Add seasonal variation
for y in range(years):
    for l in range(locations):
        # Add seasonal cycle (higher in summer, lower in winter)
        temp_data[y, l, :] += 10 * np.sin(np.linspace(0, 2*np.pi, measurements_per_year))
        # Add location-specific offset
        temp_data[y, l, :] += l * 2  # Each location is 2 degrees warmer

print("Sample temperature data created")
print(f"Shape: {temp_data.shape} (years, locations, measurements)")

# 1. Calculate the mean, standard deviation, and variance of the temperature data
print("\n1. Mean, standard deviation, and variance of temperature data:")
temp_mean = np.mean(temp_data)
temp_std = np.std(temp_data)
temp_var = np.var(temp_data)

print(f"Overall mean temperature: {temp_mean:.2f}°C")
print(f"Overall standard deviation: {temp_std:.2f}°C")
print(f"Overall variance: {temp_var:.2f}°C²")

# By location
location_means = np.mean(temp_data, axis=(0, 2))
print("\nMean temperature by location:")
for loc, mean in enumerate(location_means):
    print(f"Location {loc+1}: {mean:.2f}°C")

# 2. Identify the location with the highest and lowest average temperature
print("\n2. Location with highest and lowest average temperature:")
max_temp_loc = np.argmax(location_means)
min_temp_loc = np.argmin(location_means)

print(f"Location with highest average temp: Location {max_temp_loc+1} ({location_means[max_temp_loc]:.2f}°C)")
print(f"Location with lowest average temp: Location {min_temp_loc+1} ({location_means[min_temp_loc]:.2f}°C)")

# 3. Determine the year with the highest temperature variance
print("\n3. Year with highest temperature variance:")
year_variances = np.var(temp_data, axis=(1, 2))

max_var_year = np.argmax(year_variances)
print(f"Year with highest temperature variance: Year {max_var_year+1} ({year_variances[max_var_year]:.2f}°C²)")

print("\n--- ADDITIONAL NUMPY OPERATIONS ---\n")

# Boolean indexing
print("Boolean indexing example:")
# Create a sample array
sample_array = np.array([1, 5, 3, 8, 2, 7, 9, 4, 6])
condition = sample_array > 5
filtered_array = sample_array[condition]

print(f"Original array: {sample_array}")
print(f"Condition (> 5): {condition}")
print(f"Filtered array: {filtered_array}")

# Reshape a 1D array into a 2D array
print("\nReshaping a 1D array into a 2D array:")
one_d_array = np.arange(12)  # [0, 1, 2, ..., 11]
# Reshape to 3x4 matrix
reshaped_array = one_d_array.reshape(3, 4)

print(f"1D array: {one_d_array}")
print("Reshaped to 3x4 array:")
print(reshaped_array)

# Stack arrays horizontally and vertically
print("\nStacking arrays horizontally and vertically:")
array1 = np.array([[1, 2], [3, 4]])
array2 = np.array([[5, 6], [7, 8]])

print("Array 1:")
print(array1)
print("Array 2:")
print(array2)

# Horizontal stacking (along columns)
h_stack = np.hstack((array1, array2))
print("Horizontal stack (hstack):")
print(h_stack)

# Vertical stacking (along rows)
v_stack = np.vstack((array1, array2))
print("Vertical stack (vstack):")
print(v_stack)

# Find max and min values with their indexes
print("\nFinding max and min values with their indexes:")
sample_2d = np.array([[3, 7, 2], [8, 1, 5], [4, 9, 6]])

print("2D array:")
print(sample_2d)

max_val = np.max(sample_2d)
min_val = np.min(sample_2d)

# Find the indices of the max and min values
max_idx = np.unravel_index(np.argmax(sample_2d), sample_2d.shape)
min_idx = np.unravel_index(np.argmin(sample_2d), sample_2d.shape)

print(f"Maximum value: {max_val} at position {max_idx}")
print(f"Minimum value: {min_val} at position {min_idx}")

# Calculate sum along rows and columns
print("\nCalculating sum along rows and columns:")
row_sums = np.sum(sample_2d, axis=1)
col_sums = np.sum(sample_2d, axis=0)

print("Original 2D array:")
print(sample_2d)
print(f"Row sums: {row_sums}")
print(f"Column sums: {col_sums}")
```
