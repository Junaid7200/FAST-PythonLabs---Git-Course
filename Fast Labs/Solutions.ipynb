```python
# Input: Number of units consumed
units = float(input("Enter the number of electricity units consumed: "))

# Initialize bill amount
bill = 0

# Calculate bill based on given conditions
if units <= 50: # if statements are very simple, you write the condition right after the if keyword with no () and you put a colon to start the if block
    bill = units * 0.50
elif units <= 150:  # else if is called elif in python
    bill = (50 * 0.50) + ((units - 50) * 0.75)
elif units <= 250:
    bill = (50 * 0.50) + (100 * 0.75) + ((units - 150) * 1.20)
else:   # else block required no conditions obv
    bill = (50 * 0.50) + (100 * 0.75) + (100 * 1.20) + ((units - 250) * 1.50)

# Add 20% surcharge
surcharge = bill * 0.20
total_bill = bill + surcharge

# Display the results
print(f"Electricity Bill before surcharge: Rs. {bill:.2f}")
print(f"Surcharge (20%): Rs. {surcharge:.2f}")
print(f"Total Electricity Bill: Rs. {total_bill:.2f}")
```

```python
# Function to check if a number is perfect (a positive number that is equal to the sum of all its divisors, for instance: 6 has divisors 1, 2, 3, so 6 = 1 + 2 + 3 = 6; hence 6 is perfect)
def is_perfect(num):    # we use def keyword to start a function in python, after def, you put the name of the function and the parameters in ()
    sum_of_divisors = 0
    for i in range(1, num): # a for loop starts with the for keyword, then the variable to iterate with, then in keyword, and then the range of we have to iterate through. the range keyword doesn't iterate on the end value, it iterates on the value before the end value, so if its 1, 10...it would actually be 1, 2, 3, 4, 5, 6, 7, 8, 9
        if num % i == 0:    # % is used for remainder as usual
            sum_of_divisors += i    # += is add and assign as usual
    return sum_of_divisors == num   # thats a == not = btw, so a boolean will be returned from the function

# Input range
start = int(input("Enter the starting range: "))    # int typecasts the input to int type
end = int(input("Enter the ending range: "))

# Finding perfect numbers within the range
perfect_numbers = [num for num in range(start, end + 1) if is_perfect(num)]    # list comprehension: this is kinda advanced but understandable, we can write for loops in a single line in python, what this does is as follows:

"""
for num (iterator) in the range(start, end + 1), the range is as usual, explained in the above for loop. if the num in the range is perfect (true or false returned from function), then that number will be added to the perfect_numbers list. the two num is the confusing part mainly, the first num is the final value we add to perfect_numbers list if the function returns true, and the second num is the iterator that is iterating through that range
"""

# Display the results
print("The Perfect numbers within the given range:", *perfect_numbers)
```

```python
# using list comprehensions on lists and extracting elements from them
# Given lists
list1 = [10, 20, 25, 30, 35]
list2 = [40, 45, 60, 75, 90]

# Extract odd numbers from list1
odd_numbers = [num for num in list1 if num % 2 != 0]

# Extract even numbers from list2
even_numbers = [num for num in list2 if num % 2 == 0]

# Combine both lists
result_list = odd_numbers + even_numbers    # we can simply combine 2 lists and make a third list from it

# Display the result
print("result list:", result_list)
```

```python
# inner and outer functions, this was kinda covered in memoization, the outer function would basically have the inner function initialized and then the inner function would be called in the outer function

def outer_function(a, b):
    # Inner function to add a and b being created inside outer function
    def inner_function():
        return a + b
    
    # Call inner function and add 5
    result = inner_function() + 5   # the inner function is called inside the outer function
    return result

# Example usage
print(outer_function(3, 7))  # Output: 15   # the outer function is simply called in global scope like normal
```

```python
import random   # we can use import keyword to import libraries/modules into our python environment

# User input for list size
size = int(input("Enter the size of the list: "))

# Generate a list of random numbers
random_list = [random.randint(1, 100) for _ in range(size)] # random.randint(1, 100) uses dot notation, so first we mention the module name (which is random), then we put a dot and call the function from the module (randint()), after that we give the function parameters which were 1 and 100 (what randint does is simply provide a random number from a given range, which was 1 to 100 in this case). the _ is just a conventional iterator we use to iterate through a range

print("Generated list:", random_list)

# Find min and max without using min() and max()
smallest = random_list[0]
largest = random_list[0]

for num in random_list:
    if num < smallest:
        smallest = num
    if num > largest:
        largest = num

# Display the results
print(f"Minimum value: {smallest}")
print(f"Maximum value: {largest}")
```

```python
def sum_lists(list1, list2):
    # Find the length of the shorter list
    min_length = min(len(list1), len(list2))

    # Sum elements up to the shorter list length
    list3 = [list1[i] + list2[i] for i in range(min_length)]

    # Add remaining elements from the longer list (if any)
    list3.extend(list1[min_length:] or list2[min_length:])
    
    return list3

# Given lists
list1 = [11, 22, 33, 44, 21, 54, 67, 54, 33, 222, 4]
list2 = [3, 4, 5, 32, 21, 33, 66, 75, 87, 97, 1]

# Function call
result_list = sum_lists(list1, list2)
print("Summed List:", result_list)
```

```python
# Library Catalog System

# Dictionary to store books (book_id as key and details as value)
library_catalog = {}

# Function to add a new book
def add_book(book_id, title, author, publication_year):
    library_catalog[book_id] = {
        "Title": title,
        "Author": author,
        "Publication Year": publication_year,
        "Available": True  # Book is available by default
    }
    print(f"Book '{title}' added successfully.")

# Function to update book information
def update_book(book_id, title=None, author=None, publication_year=None):
    if book_id in library_catalog:
        if title:
            library_catalog[book_id]["Title"] = title
        if author:
            library_catalog[book_id]["Author"] = author
        if publication_year:
            library_catalog[book_id]["Publication Year"] = publication_year
        print("Book details updated successfully.")
    else:
        print("Book ID not found.")

# Function to search for books by title or author
def search_books(query):
    results = [
        details for book_id, details in library_catalog.items()
        if query.lower() in details["Title"].lower() or query.lower() in details["Author"].lower()
    ]
    if results:
        print("Search Results:")
        for book in results:
            print(book)
    else:
        print("No books found.")

# Function to borrow a book
def borrow_book(book_id):
    if book_id in library_catalog:
        if library_catalog[book_id]["Available"]:
            library_catalog[book_id]["Available"] = False
            print(f"You have borrowed '{library_catalog[book_id]['Title']}'.")
        else:
            print("Book is already borrowed.")
    else:
        print("Book ID not found.")

# Function to return a book
def return_book(book_id):
    if book_id in library_catalog:
        if not library_catalog[book_id]["Available"]:
            library_catalog[book_id]["Available"] = True
            print(f"You have returned '{library_catalog[book_id]['Title']}'.")
        else:
            print("This book was not borrowed.")
    else:
        print("Book ID not found.")

# Function to display catalog report
def catalog_report():
    print("\nLibrary Catalog Report:")
    for book_id, details in library_catalog.items():
        print(f"ID: {book_id}, Title: {details['Title']}, Author: {details['Author']}, Year: {details['Publication Year']}, Available: {'Yes' if details['Available'] else 'No'}")

# Sample usage
add_book(1, "1984", "George Orwell", 1949)
add_book(2, "To Kill a Mockingbird", "Harper Lee", 1960)
add_book(3, "The Great Gatsby", "F. Scott Fitzgerald", 1925)

update_book(2, author="Harper Lee (Updated)")

search_books("1984")
borrow_book(1)
catalog_report()

return_book(1)
catalog_report()
```
