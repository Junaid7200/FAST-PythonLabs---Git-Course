```python
import string
import random as rand

class PasswordGenerator:
    def __init__(self, length=12):
        self.length = length
        self.character_set = [string.ascii_lowercase, string.ascii_uppercase, string.digits, string.punctuation]
    def set_length(self, length):
        self.length = length
    def set_character_set(self, character_sets_name):
        self.character_set = []
        for name in character_sets_name:
            if name == 'lowercase':
                self.character_set.append(string.ascii_lowercase)
            elif name == 'uppercase':
                self.character_set.append(string.ascii_uppercase)
            elif name == 'digits':
                self.character_set.append(string.digits)
            elif name == 'punctuation':
                self.character_set.append(string.punctuation)
            else:
                print("Invalid character set name")
                return
    def generate_password(self):
        password = ''
        for i in range(self.length):
            character_set = rand.choice(self.character_set)
            password += rand.choice(character_set)
        print("the generated password is: ", password)


p_generator = PasswordGenerator()
p_generator.generate_password()
p_generator.set_length(10)
p_generator.generate_password()
p_generator.set_character_set(['lowercase', 'uppercase', 'digits'])
p_generator.generate_password()
```

```python
import string
from collections import Counter

class TextAnalyzer:
    def __init__(self, file_path):
        self.file_path = file_path
        self.word_count: dict[str, int] = {}

    def read_file(self):
        with open(self.file_path, 'r') as f:
            for line in f:
                # Remove punctuation using translate
                line = line.translate(str.maketrans('', '', string.punctuation))
                for word in line.split():
                    word = word.lower()  # Convert word to lowercase
                    if word in self.word_count:
                        self.word_count[word] += 1
                    else:
                        self.word_count[word] = 1
        return self.word_count

    def get_top_words(self, n):
        return Counter(self.word_count).most_common(n)

    def display_top_words(self, n):
        top_words = self.get_top_words(n)
        print("Top", n, "words:")
        for word, count in top_words:
            print(word, ":", count)


text_analyzer = TextAnalyzer("text.txt")
text_analyzer.read_file()
text_analyzer.display_top_words(5)
```

```python
import random
class HangmanGame:
    def __init__(self, word_list=["Vocabulary", "forms", "essential", "part", "every", "competitive", "exam", "units", "universe", "world", "compilation"], attempts=6):
        self.attempts = attempts
        self.word_list = word_list
        self.guessed_letters = []
        self.is_won = False
        self.is_lost = False
        self.word_to_guess = random.choice(self.word_list)
        self.combined = ""
    def display_word(self):
        print(f"the word is: {self.word_to_guess}")
        for letter in self.word_to_guess:
            if letter in self.guessed_letters:
                print(letter, end=" ")
            else:
                print("_", end=" ")
        print()
    def make_guess(self):
        print(f"The letters you have guessed so far are: {self.guessed_letters}\nAttempts left: {self.attempts}")
        guess = input("Enter a letter: ")
        if guess in self.guessed_letters:
            print("You already guessed this letter.")
            return
        self.guessed_letters.append(guess)
        if guess in self.word_to_guess:
            self.combined += guess
            print("Correct guess!")
        else:
            print("Incorrect guess.")
            self.attempts -= 1
        if self.attempts == 0:
            self.is_lost = True
        if self.combined == self.word_to_guess:
            self.is_won = True
    def is_game_over(self):
        if self.is_won:
            print("You have won the game!")
            return True
        elif self.is_lost:
            print("You have lost the game.")
            return True
        else:
            return False
    def play(self):
        while not self.is_game_over():
            self.display_word()
            self.make_guess()
            print(f"the combined word is: {self.combined}")

game = HangmanGame()
game.play()
```

```python
class ListOperations:
    def __init__(self, list=None):
        self.my_list = list if list is not None else []
    def add(self, element):
        self.my_list.append(element)
    def remove(self, element):
        self.my_list.remove(element)
    def find_max(self):
        return max(self.my_list)
    def find_min(self):
        return min(self.my_list)
    def display(self):
        print(self.my_list)

list_operations = ListOperations([1, 2, 3, 4, 5])
list_operations.display()
list_operations.add(6)
list_operations.display()
list_operations.remove(3)
list_operations.display()
print("Maximum element:", list_operations.find_max())
print("Minimum element:", list_operations.find_min())
```

```python
import random as rand
class NumberGuessingGame:
    def __init__(self, max_number=100, max_attempts=None, min_number=1):
        self.random_number = rand.randint(min_number, max_number)
        self.max_attempts = rand.randint(1, max_number)if max_attempts is None else max_attempts
        self.attempts_left = self.max_attempts
        self.is_won = False
    def guess_number(self, user_guess):
        if user_guess == self.random_number:
            self.is_won = True
            print("You have won the game!")
            return True
        elif user_guess < self.random_number:
            print("Too low. Try again.")
            self.attempts_left -= 1
            return False
        elif user_guess > self.random_number:
            print("Too high. Try again.")
            self.attempts_left -= 1
            return False
        if self.attempts_left == 0:
            print("You have lost the game.")
            return True
        return False
    def is_game_over(self, guess):
        if guess == -1:
            return False
        if self.is_won:
            return True
        elif self.attempts_left == 0:
            return True
        else:
            return False
    def play(self):
        guess = -1
        while not self.is_game_over(guess):
            print(f"attempts: {self.attempts_left}")
            guess = int(input("Enter your guess"))
            print(f"your guess is : {guess}")
            self.guess_number(guess)


game = NumberGuessingGame()
game.play()
```
